```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/F1Drivers_Dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

Visualisasi Perbandingan Jumlah Peserta dan Jumlah Juara Dunia F1

```python
sns.countplot(data=df, x='Champion', palette='coolwarm')
plt.title("Perbandingan Jumlah Peserta dan Jumlah Juara Dunia F1")
plt.xticks([0, 1], ['Peserta', 'Juara'])
plt.show()
```

Visualisasi Perbandingan Jumlah Balapan dengan Jumlah Poin

```python
sns.scatterplot(data=df, x='Race_Entries', y='Points', hue='Champion')
plt.title("Jumlah Balapan vs Poin")
plt.xlabel("Jumlah Balapan")
plt.ylabel("Total Poin")
plt.show()
```

Visualisasi Distribusi Negara Asal Pembalap F1

```python
df['Nationality'].value_counts().head(10).plot(kind='barh', figsize=(10,6), color='lightgreen')
plt.title("Negara Teratas Asal Pembalap F1")
plt.xlabel("Jumlah Pembalap")
plt.gca().invert_yaxis()
plt.show()
```

Visualisasi Distribusi Poin: Juara Dunia vs Non-Juara

```python
sns.violinplot(data=df, x="Champion", y="Points", palette="pastel")
plt.title("Distribusi Poin: Juara Dunia vs Non-Juara")
plt.xticks([0, 1], ['Bukan Juara', 'Juara'])
plt.show()
```

Visualisasi Korelasi Antara Statistik Performa

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Points', 'Race_Entries', 'Race_Wins', 'Podiums', 'Pole_Positions', 'Win_Rate']].corr(), annot=True, cmap='Blues')
plt.title("Korelasi Antara Statistik Performa")
plt.show()
```

**Insight yang saya dapatkan dari visualisasi data diatas, yaitu:**
1. Terdapat ketimpangan besar antara pembalap juara dan non-juara.
2. Juara dunia memiliki jumlah balapan dan poin jauh lebih besar.
3. Ada pembalap dengan banyak balapan namun poin rendah yang berarti partisipasi pembalap tersebut lama tapi tidak kompetitif.
4. Pembalap F1 sangat didominasi oleh non-juara dunia (hanya ~10% yang juara).
5. Negara dengan pembalap terbanyak, yaitu: Inggris, Italia, Jerman.
6. Terdapat korelasi kuat antara kemenangan, podium, dan poin.
6. Win rate tinggi sangat jarang, mayoritas pembalap memiliki rasio kemenangan rendah.
